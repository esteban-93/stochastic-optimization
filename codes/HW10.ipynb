#### *El granjero Ted tiene ahora nuevas cosas dentro de su problema de optimización, ya que esta enfrentando incertidumbres adicionales. Sea $\Gamma$ el conjunto de productos que tiene que gestionar, para este caso $\Gamma$ = \{Wheat, corn, beets}*.
#### *La demanda del ganado es satisfecha por su producción y dado el caso por lo que compra $y$. La cantidad de producción local es definida por la cantidad de tierra $x$ destinada para cada producto. Él puede vender producción extra, la primera parte extra de su producción $z^{I}$ la vende a un precio alto, si la producción es realmente alta, él puede vender el exceso de producción $z^{II}$ en un precio más bajo.*
#### *El granjero quiere maximizar las ganancias netas vendiendo producto extra pero cubriendo los precios de compra y venta.*

#### *A continuación se muestra una versión general de su programa lineal:*

#### $$\underset{x,y,z}{minimize} \sum_{i \in \Gamma} \left (c_i x_i + \widetilde{q}_i y_i - \widetilde{h}_i^I z_i^I - h_i^{II} z_i^{II} \right ) $$

#### *Sujeto a:*
#### $$\sum_{i \in \Gamma} x_i \leq l$$
#### $$ \widetilde{a}_i x_i + y_i - z_i^I - z^{II} \geq \widetilde{d}_i \quad \ \forall \ i \in \Gamma $$
#### $$z_i^I \leq b_i, \quad \forall \ i \in \Gamma$$
#### $$x_i,\,y_i,\,z_i^I,\,z_i^{II} \geq 0, \quad \forall \ i \in \Gamma$$


#### *Los costos de compra $\widetilde{q}$ están en $\frac{$}{Ton}$, precio alto $\widetilde{h}^I$ en $\frac{$}{Ton}$, los rendimientos $\widetilde{a}$ en $\frac{Ton}{acre}$, y la demanda de producto $\widetilde{d}$ tienen incertidumbre de acuerdo a las siguientes consideraciones:*

### <font color="Red" size="6"> problema 1</font>
#***Construya la contra parte robusta (RC) del problema lineal del granjero con incertidumbre considerando la incertidumbre Manhattan para el conjunto de datos. $\zeta = [\zeta_{1},\, \zeta_{2}, \, ..., \zeta_{11}]$ es tal que $\|\zeta\|_{1} \leq 1$***

#### *Se tienen incertidumbres en la función objetivo y en la restricción de demanda y materialización de producto. Hay que mover digamos de cierta manera esa incertidumbre de la F.O hacia las restricciones, para incorporar ese parámetro dentro del conjunto $\mathcal{U}$.*

#### *Rescribimos la F.O empleando el concepto del epigrafo*

#### $$\underset{x,y,z,t}{minimize} \,\, t$$

#### *Sujeto a:*
#### $$\sum_{i \in \Gamma} \left (c_i x_i + \widetilde{q}_i y_i - \widetilde{h}_i^I z_i^I - h_i^{II} z_i^{II} \right ) \leq t $$
#### $$\sum_{i \in \Gamma} x_i \leq l$$
#### $$ \widetilde{a}_i x_i + y_i - z_i^I - z^{II} \geq \widetilde{d}_i \quad \ \forall \ i \in \Gamma $$
#### $$z_i^I \leq b_i, \quad \forall \ i \in \Gamma$$
#### $$x_i,\,y_i,\,z_i^I,\,z_i^{II} \geq 0, \quad \forall \ i \in \Gamma$$

<br>

#### *El conjunto $\mathcal{U}$ se puede escribir de la siguiente forma:*

$$\mathcal{U} = \begin{pmatrix} c^{\top}_o & d_o\\ A_o & b_o \\ \end{pmatrix} + \sum_{l=1}^{L} \zeta_{l} \begin{pmatrix} c^{\top}_{l} & d_{l} \\ A_{l} & b_{l}\end{pmatrix} \, : \zeta \in Z \subset \Re^{L}$$

#### *Para nuestro problema será:*
$$\mathcal{U} = \begin{pmatrix} q_{w}\\ q_{c}\\ q_b \\ h^{I}_w \\h^{I}_c \\h^{I}_{b}\\a_w\\a_c\\a_b\\d_w \\d_c\\d_b \end{pmatrix} = \begin{pmatrix} q^{o}_w = 238\\q^{o}_c = 210\\ q^{o}_b = 100 \\ h^{I}_w = 170\\h^{I}_c = 150\\ h^{I}_b = 36\\ a_w = 2.5\\ a_c = 3\\ a_b = 20 \\ d_w = 200\\  d_c = 240\\  d_b = 0 \end{pmatrix} + \eta \sum_{l=1}^{11} \zeta_{l} \begin{pmatrix} q^{0}_w & 0, ..., &0\\0 & q^{o}_c, ..., &0\\ ..., &,... &,...\\0 & 0, ..., & d_{b} \end{pmatrix}$$

<br>

####*Para considerar la incertidumbre tipo Manhattan, tomamos lo de clase:*
$$ (a^{o})^{\top}x + \mathcal{u} \leq b^{o}$$
$$ (a^{l})^{\top}x - b^{l} \leq \mathcal{u}, \quad \forall l=1, ..., L $$
$$  b^{l} - (a^{l})^{\top}x \leq \mathcal{u}, \quad \forall l=1, ..., L $$

<br>

#### *Hay que convertir la F.O que tiene incertidumbre y la restricción de demanda, las otras quedan igual.*
$$\sum_{i \in \Gamma} \left (c_i x_i + q^{o}_i y_i - (h^{I}_i)^o z_i^I - h_i^{II} z_i^{II} \right ) + \mathcal{u_{1}} \leq t $$

#### *Se ponen las restricciones asociadas a esas variables inciertas:*

$$\eta q^{o}_{i}y_{i} \leq \mathcal{u_{1}} \quad \forall i \in \Gamma$$
$$-\eta q^{o}_{i}y_{i} \leq \mathcal{u_{1}} \quad \forall i \in \Gamma$$
$$\eta (h^{I}_i)^oz^{I}_{i} \leq \mathcal{u_{1}} \quad \forall i \in \Gamma $$
$$-\eta (h^{I}_i)^oz^{I}_{i} \leq \mathcal{u_{1}} \quad \forall i \in \Gamma $$

#### *Para la siguiente restricción se tiene:*
#### *La cambiamos de sentido*

$$-a^{o}_i x_i - y_i + z_i^I + z^{II} + \mathcal{u_{2}} \leq -d^{o}_i \quad \ \forall \ i \in \Gamma$$

$$\eta a^{o}_{i}x_{i} \leq \mathcal{u_{2}} \quad \forall i \in \Gamma$$
$$-\eta a^{o}_{i}x_{i} \leq \mathcal{u_{2}} \quad \forall i \in \Gamma$$
$$-\eta d^{o}_i \leq \mathcal{u_{2}} \quad \forall i \in \Gamma $$
$$\eta d^{o}_i \leq \mathcal{u_{2}} \quad \forall i \in \Gamma $$

```python
import pandas as pd
import pyomo.environ as pyo
from pyomo.util.infeasible import log_infeasible_constraints
from pyomo.opt import SolverStatus,TerminationCondition
import numpy as np
import math

m = pyo.AbstractModel()
opt = pyo.SolverFactory("glpk", executable='/usr/bin/glpsol')

m.Z = pyo.Set(doc="Productos a sembrar")
m.Uncert = pyo.Set(doc="Conjunto de parámetros inciertos")


# Parámetros
m.c = pyo.Param(m.Z, domain=pyo.NonNegativeReals, doc="costo del producto i")
m.h2 = pyo.Param(m.Z, domain=pyo.NonNegativeReals, doc="precio de venta alto del producto i")
m.l = pyo.Param(domain=pyo.NonNegativeReals, doc="Acres disponibles")
m.b =  pyo.Param(m.Z, domain=pyo.NonNegativeReals, doc="Límites del producto i")
m.eta = pyo.Param(domain=pyo.NonNegativeReals, mutable=True, default=0.3,
                  doc="cambios para los parámetros inciertos")
m.rho = pyo.Param(domain=pyo.Reals, mutable=True, default=-0.9,
                  doc="cambios para los parámetros inciertos")
m.nominals = pyo.Param(m.Uncert, m.Z, domain=pyo.Reals, doc="Valores nominales", default=0)


#Vars
m.X = pyo.Var(m.Z, domain=pyo.NonNegativeReals, doc="Cantidad de tierra")
m.Y = pyo.Var(m.Z, domain=pyo.NonNegativeReals,doc="Compras")
m.z1 = pyo.Var(m.Z, domain=pyo.NonNegativeReals,doc="Producto i vendido a alto precio")
m.z2 = pyo.Var(m.Z, domain=pyo.NonNegativeReals,doc="Producto i vendido a precio bajo")

#Vars of the uncertain part
m.U1 = pyo.Var(domain=pyo.NonNegativeReals, doc="")
m.U2 = pyo.Var(m.Z, domain=pyo.NonNegativeReals, doc="")
m.t = pyo.Var(domain=pyo.Reals, doc="epigrafo variable")

# F.O
def objective(m):
    return m.t
m.Cost = pyo.Objective(rule=objective, sense=pyo.minimize)

# Restricción incierta de la FO
def constraint1(m):
    return sum(m.c[i]*m.X[i] + m.nominals["q", i]*m.Y[i] - m.nominals["h1", i]*m.z1[i] -
               m.h2[i]*m.z2[i] + m.U1 for i in m.Z) <= m.t
m.ObjFuncConst = pyo.Constraint(rule=constraint1, doc="Restricción de la FO")

def constraint8(m, i):
    return m.eta*m.nominals["q", i]*m.Y[i] <= m.U1
m.ObjFuncConst1 = pyo.Constraint(m.Z, rule=constraint8, doc="Restricción FO 1")

def constraint9(m, i):
    return - m.eta*m.nominals["q", i]*m.Y[i] <= m.U1
m.ObjFuncConst2 = pyo.Constraint(m.Z, rule=constraint9, doc="Restricción FO 2")

def constraint10(m, i):
    return m.eta*m.nominals["h1", i]*m.z1[i] <= m.U1
m.ObjFuncConst3 = pyo.Constraint(m.Z, rule=constraint10, doc="Restricción FO 3")

def constraint11(m, i):
    return - m.eta*m.nominals["h1", i]*m.z1[i] <= m.U1
m.ObjFuncConst4 = pyo.Constraint(m.Z, rule=constraint11, doc="Restricción FO 4")


# Restricciones del problema-2 **** Función objetivo
def constraint13(m, i):
    return m.eta*m.Y[i]*(m.nominals["q", i] + m.rho*m.nominals["a", i]) <= m.U1
m.ObjFuncConst1P2 = pyo.Constraint(m.Z, rule=constraint13, doc="Restricción FO-1 P2")

def constraint14(m, i):
    return - m.eta*m.Y[i]*(m.nominals["q", i] + m.rho*m.nominals["a", i]) <= m.U1
m.ObjFuncConst2P2 = pyo.Constraint(m.Z, rule=constraint14, doc="Restricción FO-2 P2")

def constraint15(m, i):
    return - m.eta*m.z1[i]*m.nominals["h1", i] <= m.U1
m.ObjFuncConst3P2 = pyo.Constraint(m.Z, rule=constraint15, doc="Restricción FO-3 P2")

def constraint16(m, i):
    return m.eta*m.z1[i]*m.nominals["h1", i] <= m.U1
m.ObjFuncConst4P2 = pyo.Constraint(m.Z, rule=constraint16, doc="Restricción FO-4 P2")

# Restricciones del problema-2 **** Balance de demanda
def constraint17(m, i):
    return m.eta*m.X[i]*(m.nominals["a", i] + m.rho*m.nominals["q", i]) <= m.U2[i]
m.DemandConstr1P2 = pyo.Constraint(m.Z, rule=constraint17, doc="Incertidumbre en demanda-1 P2")

def constraint18(m, i):
    return -m.eta*m.X[i]*(m.nominals["a", i] + m.rho*m.nominals["q", i]) <= m.U2[i]
m.DemandConstr2P2 = pyo.Constraint(m.Z, rule=constraint18, doc="Incertidumbre en demanda-2 P2")

def constraint19(m, i):
    return - m.eta*m.nominals["d", i] <= m.U2[i]
m.DemandConstr3P2 = pyo.Constraint(m.Z, rule=constraint19, doc="Incertidumbre en demanda-3 P2")

def constraint20(m, i):
    return m.eta*m.nominals["d", i] <= m.U2[i]
m.DemandConstr4P2 = pyo.Constraint(m.Z, rule=constraint20, doc="Incertidumbre en demanda-4 P2")


# Cantidad de acres disponible
def constraint2(m):
    return sum(m.X[i] for i in m.Z) <= m.l
m.AcresAmount = pyo.Constraint(rule=constraint2, doc="Límite de acres")

# Restricción de límite de toneladas
def constraint12(m, i):
    return m.z1[i] <= m.b[i]
m.Limit = pyo.Constraint(m.Z, rule=constraint12, doc="Restricción del toneladas")

# Restricción incierta de la demanda
def constraint3(m, i):
    return - m.nominals["a", i]*m.X[i] - m.Y[i] + m.z1[i] + m.z2[i] + m.U2[i] <= -m.nominals["d", i]
m.DemandConstr = pyo.Constraint(m.Z, rule=constraint3, doc="Cumplimiento de demanda")

def constraint4(m, i):
    return m.eta*m.nominals["a", i]*m.X[i] <= m.U2[i]
m.DemandConstr1 = pyo.Constraint(m.Z, rule=constraint4, doc="Incertidumbre en demanda 1")

def constraint5(m, i):
    return - m.eta*m.nominals["a", i]*m.X[i] <= m.U2[i]
m.DemandConstr2 = pyo.Constraint(m.Z, rule=constraint5, doc="Incertidumbre en demanda 2")

def constraint6(m, i):
    return - m.eta*m.nominals["d", i] <= m.U2[i]
m.DemandConstr3 = pyo.Constraint(m.Z, rule=constraint6, doc="Incertidumbre en demanda 3")

def constraint7(m, i):
    return m.eta*m.nominals["d", i] <= m.U2[i]
m.DemandConstr4 = pyo.Constraint(m.Z, rule=constraint7, doc="Incertidumbre en demanda 4")
```
